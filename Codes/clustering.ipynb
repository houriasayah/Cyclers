```python
import  pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Load the dataset
trip = pd.read_csv("trip_subtrip.csv", sep=";")
```

```python
    # Extract hours, minutes, and seconds from the duration_x column
trip["duration_hours"] = trip["duration_x"].str.extract(r"^(\d+):", expand=False).astype(float)
trip["duration_minutes"] = trip["duration_x"].str.extract(r":(\d+):", expand=False).str.slice(1).astype(float)
trip["duration_seconds"] = trip["duration_x"].str.extract(r":(\d+)$", expand=False).str.slice(1).astype(float)
```

```python

    # Convert the numeric columns to numeric type
trip[[ "avg_speed", "distance_x","subtrip_type"]] = trip[[ "avg_speed", "distance_x","subtrip_type"]].apply(pd.to_numeric)
```

```python

    # Combine the duration columns into a single numeric column representing the duration in seconds
trip["duration_seconds"] = trip["duration_hours"] * 3600 + trip["duration_minutes"] * 60 + trip["duration_seconds"]
```

```python
    # Remove the original duration_x column
trip = trip.drop("duration_x", axis=1)
```

```python
    # Scale the numeric columns
numeric_columns = ["duration_seconds", "avg_speed", "distance_x"]
scaled_trip = trip[numeric_columns].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python

    # Combine the categorical and scaled numeric columns
clustering_data = pd.concat([trip[["subtrip_type"]], scaled_trip], axis=1)
```

```python

    # Perform k-means clustering (you can adjust the number of clusters by changing the n_clusters parameter)
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(clustering_data)
```

```python

    # Add the cluster assignments to the original dataset
trip["cluster"] = clusters
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.impute import SimpleImputer

# Charger le dataset
trip = pd.read_csv("trip_subtrip.csv", sep=";")

# Vérifier que les colonnes requises sont présentes dans le dataset
required_columns = ["subtrip_type", "duration_x", "avg_speed", "distance_x"]
if set(required_columns).issubset(trip.columns):
    # Extraction et conversion de la durée en secondes
    # Assurez-vous que la colonne duration_x est au format correct, ex. hh:mm:ss ou mm:ss
    trip['duration_seconds'] = pd.to_timedelta(trip['duration_x']).dt.total_seconds()
    
    # Convertir les colonnes numériques
    trip[["avg_speed", "distance_x"]] = trip[["avg_speed", "distance_x"]].apply(pd.to_numeric, errors='coerce')
    
    # Encoder la colonne 'subtrip_type'
    le = LabelEncoder()
    trip['subtrip_type_encoded'] = le.fit_transform(trip['subtrip_type'])
    
    # Préparer les données pour le clustering
    numeric_columns = ["duration_seconds", "avg_speed", "distance_x", "subtrip_type_encoded"]
    
    # Imputation pour gérer les valeurs NaN
    imputer = SimpleImputer(strategy="mean")
    imputed_data = imputer.fit_transform(trip[numeric_columns])
    
    # Normalisation des données
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(imputed_data)
    
    # Clustering KMeans
    kmeans = KMeans(n_clusters=50, random_state=0)
    clusters = kmeans.fit_predict(scaled_features)
    
    # Ajouter les affectations de cluster au dataset original
    trip['cluster'] = clusters
else:
    print("Les colonnes requises ne sont pas trouvées dans le dataset.")
```
