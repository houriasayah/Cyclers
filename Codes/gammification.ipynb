```python
import pandas as pd 
trip = pd.read_csv("trip_subtrip_cohérente.csv",sep = ";")
trip.head()

```

```python
#suppresion de la colonne
trip.drop(columns="Unnamed: 0", inplace= True)
```

```python
#conversion en format durée 
trip['duration_x'] = pd.to_timedelta(trip['duration_x'])
```

```python
#calcul rapide d'indicateur basique
print("sum points : ", trip["points_x"].sum())
print("sum calories : ", trip["calories_x"].sum())
print("sum distance : ", trip["distance_x"].sum())
print("sum avg_speed : ", trip["avg_speed"].sum())
print("sum duration_y : ", trip["duration_x"].sum())
print("number of users: ", trip["id_utente_x"].nunique())

```

```python
import pandas as pd
#stockage des indicateurs dans un df 
trip_sums_and_count = {
    "sum_points": trip["points_x"].sum(),  
    "sum_calories": trip["calories_x"].sum(),  
    "sum_distance": trip["distance_x"].sum(),  
    "sum_avg_speed": trip["avg_speed"].sum(), 
    "sum_duration": trip["duration_x"].sum(),  
    "number_of_users": trip["id_utente_x"].nunique() 
}

trip_df = pd.DataFrame([trip_sums_and_count])

trip_df
```
