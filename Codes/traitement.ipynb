```python
from faker import Faker
import numpy as np
import random
import pandas as pd

# Initialisation de Faker pour la génération de timestamps aléatoires
faker = Faker()

# Nombre de lignes à générer
n_rows = 8598

# Génération des identifiants aléatoires
id_utente_x = np.random.randint(low=1e13, high=1e14, size=n_rows)
trip_id = np.random.randint(low=900000, high=1000000, size=n_rows)
subtrip_id = np.random.randint(low=400000, high=500000, size=n_rows)

# Sélection aléatoire des types de sous-trajets
subtrip_types = ['Biking', 'Walking','BikeElectric','CarElectric','Metro','Train','Bus','Scooter','MotoElectric']
subtrip_type = np.random.choice(subtrip_types, size=n_rows)

# Génération des distances aléatoires
distance_x = np.random.uniform(low=0.5, high=10, size=n_rows)
distance_y = distance_x  # Pour simplifier, on prend la même valeur que distance_x

# Génération des durées aléatoires en secondes puis conversion en hh:mm:ss
duration_seconds_x = np.random.randint(low=60, high=7200, size=n_rows)
duration_x = pd.to_timedelta(duration_seconds_x, unit='s').astype(str).str.split().str[-1]
duration_y = duration_x  # Pour simplifier, on prend la même valeur que duration_x

# Génération des timestamps aléatoires
def generate_random_timestamps(n):
    start_times = [faker.date_time_between(start_date="-2y", end_date="now") for _ in range(n)]
    end_times = [start + pd.Timedelta(seconds=random.randint(60, 7200)) for start in start_times]
    return start_times, end_times

start_time_x, end_time_x = generate_random_timestamps(n_rows)
start_time_y, end_time_y = start_time_x, end_time_x  # Pour simplifier, on reprend les mêmes horaires

# Création du DataFrame
random_data = pd.DataFrame({
    'id_utente_x': id_utente_x,
    'trip_id': trip_id,
    'subtrip_id': subtrip_id,
    'subtrip_type': subtrip_type,
    'distance_x': distance_x,
    'duration_x': duration_x,
    'start_time_x': start_time_x,
    'end_time_x': end_time_x,
    'distance_y': distance_y,
    'duration_y': duration_y,
    'start_time_y': start_time_y,
    'end_time_y': end_time_y
})

# Ajustement du format des timestamps pour inclure la timezone +00:00
random_data['start_time_x'] = random_data['start_time_x'].dt.strftime('%Y-%m-%d %H:%M:%S+00:00')
random_data['end_time_x'] = random_data['end_time_x'].dt.strftime('%Y-%m-%d %H:%M:%S+00:00')
random_data['start_time_y'] = random_data['start_time_y'].dt.strftime('%Y-%m-%d %H:%M:%S+00:00')
random_data['end_time_y'] = random_data['end_time_y'].dt.strftime('%Y-%m-%d %H:%M:%S+00:00')

# Affichage des premières lignes pour vérification
random_data.head(), random_data.dtypes
```

```python
from faker import Faker
import numpy as np
import random
import pandas as pd

# Initialisation de Faker pour la génération de timestamps aléatoires
faker = Faker()

# Nombre de lignes à générer et nombre d'utilisateurs uniques
n_rows = 8598
n_users = 79

# Génération de 79 identifiants d'utilisateurs uniques et réutilisation dans les observations
unique_user_ids = np.random.randint(low=1e13, high=1e14, size=n_users)
id_utente_x = np.random.choice(unique_user_ids, size=n_rows)

# Génération des identifiants aléatoires pour les trajets
trip_id = np.random.randint(low=900000, high=1000000, size=n_rows)
subtrip_id = np.random.randint(low=400000, high=500000, size=n_rows)

# Sélection aléatoire des types de sous-trajets
subtrip_types = ['Biking', 'Walking', 'BikeElectric', 'CarElectric', 'Metro', 'Train', 'Bus', 'Scooter', 'MotoElectric']
subtrip_type = np.random.choice(subtrip_types, size=n_rows)

# Génération des distances et des durées aléatoires
distance_x = np.random.uniform(low=0.5, high=10, size=n_rows)
duration_seconds_x = np.random.randint(low=60, high=7200, size=n_rows)
duration_x = pd.to_timedelta(duration_seconds_x, unit='s').astype(str).str.split().str[-1]

# Utilisation des mêmes valeurs pour simplifier
distance_y = distance_x
duration_y = duration_x

# Génération des timestamps aléatoires
start_time_x, end_time_x = generate_random_timestamps(n_rows)
start_time_y, end_time_y = start_time_x, end_time_x  # Réutilisation des mêmes horaires

# Création du DataFrame
random_data = pd.DataFrame({
    'id_utente_x': id_utente_x,
    'trip_id': trip_id,
    'subtrip_id': subtrip_id,
    'subtrip_type': subtrip_type,
    'distance_x': distance_x,
    'duration_x': duration_x,
    'start_time_x': [start.strftime('%Y-%m-%d %H:%M:%S+00:00') for start in start_time_x],
    'end_time_x': [end.strftime('%Y-%m-%d %H:%M:%S+00:00') for end in end_time_x],
    'distance_y': distance_y,
    'duration_y': duration_y,
    'start_time_y': [start.strftime('%Y-%m-%d %H:%M:%S+00:00') for start in start_time_y],
    'end_time_y': [end.strftime('%Y-%m-%d %H:%M:%S+00:00') for end in end_time_y]
})

# Affichage des premières lignes pour vérification
random_data.head()
```
